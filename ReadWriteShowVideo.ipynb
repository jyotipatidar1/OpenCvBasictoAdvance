```python
import cv2
```

```python
cap = cv2.VideoCapture(1);
while(True):
    ret , frame = cap.read()

    
#gray scale convertion of input
#     gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
#     cv2.imshow('frame',gray)
    cv2.imshow('frame',frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
cap1 = cv2.VideoCapture("vtest.avi");

print(cap1.get(cv2.CAP_PROP_FRAME_WIDTH))
print(cap1.get(cv2.CAP_PROP_FRAME_HEIGHT))

while(cap1.isOpened()):
    ret , frame = cap1.read()
    cv2.imshow('frame',frame)
    
    if cv2.waitKey(1) & 0xFF == ord('e'):
        break
cap1.release()
cv2.destroyAllWindows()
```

```python
cap2 = cv2.VideoCapture(0);
fourcc = cv2.VideoWriter_fourcc(*'XVID');
out = cv2.VideoWriter('output.mp4',fourcc,20.0,(640,480))#20 frame/per second
while(cap2.isOpened):
    ret , frame = cap2.read()
    if ret==True:
        out.write(frame)
        cv2.imshow('frame',frame)
        print(cap2.get(cv2.CAP_PROP_FRAME_WIDTH))
        print(cap2.get(cv2.CAP_PROP_FRAME_HEIGHT))
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break
cap2.release();
out.release();
cv2.destroyAllWindows()
        
```

```python
cap1 = cv2.VideoCapture(1);

print(cap1.get(cv2.CAP_PROP_FRAME_WIDTH))#cv2.CAP_PROP_FRAME_WIDTH can be return as(3)
print(cap1.get(cv2.CAP_PROP_FRAME_HEIGHT))

cap1.set(3,1209)#set the width (if width is less than min resolution it automatically take min and if it more than max it automatically take maximum)
cap1.set(4,1350)#same like above

print(cap1.get(3))
print(cap1.get(4))

while(cap1.isOpened()):
    ret , frame = cap1.read()
    cv2.imshow('frame',frame)
    
    if cv2.waitKey(1) & 0xFF == ord('e'):
        break
cap1.release()
cv2.destroyAllWindows()
```

```python
import datetime

cap = cv2.VideoCapture(1)

print(cap.get(3))
print(cap.get(4))

while (cap.isOpened()):
    ret , frame = cap.read()
    
    font = cv2.FONT_HERSHEY_SIMPLEX
    text = 'Width: ' + str(cap.get(3)) + ' Height: ' + str(cap.get(4))
    datet = str(datetime.datetime.now())
    
    frame = cv2.putText(frame,datet,(10,50),font,1,(250,250,250),2,cv2.LINE_AA)
    cv2.imshow('frame',frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```
